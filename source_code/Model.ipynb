```python
import os
import pandas as pd
import numpy as np
import cv2
```

```python
# Directory path where images are stored
directory = 'flower_photos/train'

# Create an empty DataFrame
train_df = pd.DataFrame(columns=['Image', 'Name'])

for filename in os.listdir(directory):
    print(directory+'/'+filename)
    if os.path.isdir(directory+'/'+filename):
        for image in os.listdir(directory+'/'+filename):
            if image.endswith('.jpg') or image.endswith('.png'):
                print(directory+filename+image)
                mat = cv2.imread(directory+'/'+filename+"/"+image,cv2.IMREAD_GRAYSCALE)
                mat= cv2.resize(mat,(500,500))
                mat = cv2.Canny(mat, threshold1=30, threshold2=100)
                temp_df = pd.DataFrame({'Image': [np.array(mat)], 'Name': filename})
                train_df = train_df.append(temp_df,ignore_index=True)
train_df['Image'] = train_df['Image'].map(lambda x: x[0])
```

```python
train_df.sample(10)
```

```python
# Directory path where images are stored
directory = 'flower_photos/test'

# Create an empty DataFrame
test_df = pd.DataFrame(columns=['Image', 'Name'])

for filename in os.listdir(directory):
    print(directory+'/'+filename)
    if os.path.isdir(directory+'/'+filename):
        for image in os.listdir(directory+'/'+filename):
            if image.endswith('.jpg') or image.endswith('.png'):
                print(directory+filename+image)
                mat = cv2.imread(directory+'/'+filename+"/"+image,cv2.IMREAD_GRAYSCALE)
                mat= cv2.resize(mat,(500,500))
                mat = cv2.Canny(mat, threshold1=30, threshold2=100)
                temp_df = pd.DataFrame({'Image': [np.array(mat)], 'Name': filename})
                test_df = test_df.append(temp_df,ignore_index=True)
test_df['Image'] = test_df['Image'].map(lambda x: x[0])
```

```python
test_df.sample(10)
```

```python
from sklearn.preprocessing import LabelEncoder
model = LabelEncoder()

train_df['Name'] = model.fit_transform(train_df['Name'])
test_df['Name'] = model.transform(test_df['Name'])
```

```python
x_train = []
for image in train_df['Image']:
    x_train.append(image.flatten())
x_train = np.array(x_train)
y_train = np.array(train_df['Name'])
```

```python
x_test = []
for image in test_df['Image']:
    x_test.append(image.flatten())
x_test = np.array(x_test)
y_test = np.array(test_df['Name'])
```

### create the models

```python
def create_model_lg():
    from sklearn.linear_model import LogisticRegressionCV
    model = LogisticRegressionCV()
    model.fit(x_train, y_train)
    return model
```

```python
def create_model_knn():
    from sklearn.neighbors import KNeighborsClassifier
    model = KNeighborsClassifier(n_neighbors=5)
    model.fit(x_train, y_train)
    return model
```

```python
def create_model_svm():
    from sklearn.svm import SVC
    model = SVC()
    model.fit(x_train, y_train)
    return model
```

```python
def create_model_dt():
    from sklearn.tree import DecisionTreeClassifier
    model = DecisionTreeClassifier(max_depth=5)
    model.fit(x_train, y_train)
    return model
```

```python
def create_model_rf():
    from sklearn.ensemble import RandomForestClassifier
    model = RandomForestClassifier(n_estimators=20, max_depth=5)
    model.fit(x_train, y_train)
    return model
```

```python
def create_model_gb():
    from sklearn.ensemble import GradientBoostingClassifier
    model = GradientBoostingClassifier(n_estimators=20, max_depth=5)
    model.fit(x_train, y_train)
    return model
```

```python
def create_model_xgb():
    from xgboost import XGBClassifier
    model = XGBClassifier()
    model.fit(x_train, y_train)
    return model
```

### evaluate the model

```python
model_lg = create_model_lg()
model_knn = create_model_knn()
model_svm = create_model_svm()
model_dt = create_model_dt()
model_rf = create_model_rf()
model_gb = create_model_gb()
model_xgb = create_model_xgb()
```

```python
def evaluate_model(model):
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
    
    y_true = y_test
    y_pred = model.predict(x_test)
    
    accuracy = f"{accuracy_score(y_true, y_pred):0.2f}"
    precision = f"{precision_score(y_true, y_pred,average='macro'):0.2f}"
    recall = f"{recall_score(y_true, y_pred,average='macro'):0.2f}"
    f1 = f"{f1_score(y_true, y_pred,average='macro'):0.2f}"
    
    return accuracy, precision, recall, f1
```

```python
models = [
    {"name": "LG", "model": model_lg},
    {"name": "KNN", "model": model_knn},
    {"name": "SVM", "model": model_svm},
    {"name": "DT", "model": model_dt},
    {"name": "RF", "model": model_rf},
    {"name": "GB", "model": model_gb},
    {"name": "XGB", "model": model_xgb}
]

print('-' * 72)
print(f"|{'Algo':^10}|{'Accuracy':^15}|{'Precision':^15}|{'Recall':^15}|{'F1':^10}|")
print('-' * 72)
for model_info in models:
    evaluation = evaluate_model(model_info['model'])
    print(f"|{model_info['name']:^10}|{evaluation[0]:^15}|{evaluation[1]:^15}|{evaluation[2]:^15}|{evaluation[3]:^10}|")
    
print('-' * 72)
```
